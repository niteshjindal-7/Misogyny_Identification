```python
# Define a custom dataset class
class SpanDataset(Dataset):
    def __init__(self, embeddings, labels):
        self.embeddings = embeddings
        self.labels = labels

    def __len__(self):
        return len(self.embeddings)

    def __getitem__(self, idx):
        return self.embeddings[idx], self.labels[idx]
```

```python
i=0
KK=7

print(f'Executing for BootStrap Sample = {KK}')
train_emb1 = train_emb_all[i]
test_emb = test_emb_all[i]

    # Convert the train and test data to PyTorch tensors
train_emb1 = torch.tensor(train_emb1)
test_emb = torch.tensor(test_emb)

train_lbl1 = pd.read_csv("./PickleEmbeddings_Train_Test_Lbls/train_lbls.csv")
test_lbl = pd.read_csv("./PickleEmbeddings_Train_Test_Lbls/test_lbls.csv")
train_lbl1 = torch.tensor(train_lbl1['label'].values)
test_lbl = torch.tensor(test_lbl['label'].values)


train_emb, val_emb, train_lbl, val_lbl = train_test_split(train_emb1, 
                                                          train_lbl1, 
                                                          test_size=0.2, 
                                                          random_state=KK)

    # Create train and test datasets
train_dataset = SpanDataset(train_emb, train_lbl)
val_dataset = SpanDataset(val_emb, val_lbl)
test_dataset = SpanDataset(test_emb, test_lbl)

# Create train and test data loaders
train_dataloader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=8, shuffle=False)
test_dataloader = DataLoader(test_dataset, batch_size=8, shuffle=False)


EMBEDDING_DIM = 192
HIDDEN_SIZE = 64
OUTPUT_DIM = 1
NUM_LAYERS = 4
DROPOUT = 0.25
Epochs = 250
LEARNING_RATE = 5e-5  #0.001 #
#5e-5 

### RAHUL CODE
# class BiLSTMNet(nn.Module):
#     def __init__(self, embedding_dim, hidden_size, output_dim, num_layers, dropout):
#         super(BiLSTMNet, self).__init__()

#         self.embedding_dim = embedding_dim
#         self.hidden_size = hidden_size
#         self.num_layers = num_layers
#         self.lstm = nn.LSTM(embedding_dim, hidden_size, num_layers, bidirectional=True)
#         self.fc = nn.Linear(hidden_size*2, output_dim)
#         self.dropout = nn.Dropout(dropout)
#         self.sigmoid = nn.Sigmoid()

#     def forward(self, x):
#         output, _ = self.lstm(x)
#         output = self.fc(output)
#         output = self.sigmoid(output)
#         return output 
    
    
### NITESH CODE


class BiLSTMNet(nn.Module):
    def __init__(self, embedding_dim, hidden_size, output_dim, num_layers, dropout):
        super(BiLSTMNet, self).__init__()

        self.embedding_dim = embedding_dim
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(embedding_dim, hidden_size, num_layers, bidirectional=True, batch_first=True)
        self.fc1 = nn.Linear(hidden_size*2, 256)  # Add an additional fully connected layer
        self.relu = nn.ReLU()  # Add ReLU activation
        self.fc2 = nn.Linear(256, output_dim)  # Add another fully connected layer
        self.dropout = nn.Dropout(dropout)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        output, _ = self.lstm(x)
        output = self.fc1(output)
        output = self.relu(output)  # Apply ReLU activation
        output = self.dropout(output)  # Apply dropout
        output = self.fc2(output)
        output = self.sigmoid(output)
        return output



model = BiLSTMNet(EMBEDDING_DIM, HIDDEN_SIZE, OUTPUT_DIM, NUM_LAYERS, DROPOUT)
print(model)
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)



from sklearn.metrics import f1_score
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
criterion = criterion.to(device)

# Training Loop
for epoch in range(Epochs):
    model.train()
    running_loss_train = 0.0
    num_batches_train = 0

    # Iterate over train batches
    for inputs, labels in tqdm(train_dataloader):
        inputs = inputs.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()           # 1. Zero out the optimizer gradients
        logits = model.forward(inputs)  # 2. Forward pass on the model

        loss = criterion(logits, labels.unsqueeze(1).float()) # 3. Calculate loss
        loss.backward()                             # 4. Backprop loss
        optimizer.step()
        running_loss_train += loss.item()
        num_batches_train += 1
    train_loss = round(running_loss_train / num_batches_train, 3)
    
    # Validation phase
    model.eval()
    running_loss_val = 0.0
    num_batches_val = 0

    for inputs, labels in tqdm(val_dataloader):
        inputs = inputs.to(device)
        labels = labels.to(device)
        logits = model.forward(inputs)
        loss = criterion(logits, labels.unsqueeze(1).float())
        running_loss_val += loss.item()
        num_batches_val += 1

    val_loss = round(running_loss_val / num_batches_val, 3)
    
    if epoch % 2 == 0:
        print(f"Epoch {epoch}: Train loss {train_loss}, Val loss {val_loss}")
        
#     print(f"Epoch {epoch}: Train loss {train_loss}, Validation loss {val_loss}")        
```
