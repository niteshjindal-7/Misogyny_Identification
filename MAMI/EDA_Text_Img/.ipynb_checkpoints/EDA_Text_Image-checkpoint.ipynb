##### EDA 1.2.2 Get the POS tags and NER tags from the sentence data.

```python
def extract_text(misogyn_texts, label=True):
    if label:
        print("Texts with label 'misogyn':")
    else:
        print("Texts with label 'non_misogyn':")

    for text in misogyn_texts:
        yield text

# Load the spaCy English model
# nlp = spacy.load('en_core_web_sm')

mis_postag_and_ner_labels_list = []
for text in extract_text(misogyn_texts, label=True):
    doc = spacy_nlp(text)
    postag_and_ner_labels0 = []
    for token in doc:
        if token.ent_iob_ == 'O':
            label = 'O'
        else:
            label = token.ent_iob_ + '-' + token.ent_type_
        postag_and_ner_labels0.append((token.text, 
                                       token.tag_, 
                                       label))
    mis_postag_and_ner_labels_list.append(
        postag_and_ner_labels0)

    
non_mis_postag_and_ner_labels_list = []
for text in extract_text(non_misogyn_texts, label=False):
    doc = spacy_nlp(text)
    nonmis_postag_and_ner_labels0 = []
    for token in doc:
        if token.ent_iob_ == 'O':
            label = 'O'
        else:
            label = token.ent_iob_ + '-' + token.ent_type_
        nonmis_postag_and_ner_labels0.append((token.text, 
                                              token.tag_, 
                                              label))
    non_mis_postag_and_ner_labels_list.append(
        nonmis_postag_and_ner_labels0)

    
print('len(misogyn_texts)', len(misogyn_texts))
print('len(non_misogyn_texts)', len(non_misogyn_texts))
print('len(mis_postag_and_ner_labels_list)', 
      len(mis_postag_and_ner_labels_list))

print('len(non_mis_postag_and_ner_labels_list)', 
     len(non_mis_postag_and_ner_labels_list))

assert len(non_misogyn_texts) + len(misogyn_texts) == \
len(mis_postag_and_ner_labels_list) + \
len(non_mis_postag_and_ner_labels_list)
```

##### EDA 1.2.3 Create wordcloud for misogynous and non-misogynous 

```python
mis_postag_and_ner_labels_list
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from sklearn.feature_extraction.text import TfidfVectorizer

# List of filtered POS tags
filtered_pos_tags = ['NNP', 'NNS', 'NNPS', 'NN']

# filtered_pos_tags = ['NNP', 'NNS', 'NNPS', 'NN'
#                      'JJ', 'JJR', 'JJS', 
#                      'VB', 'VBG', 'VBN', 'VBP']

# List to store all the filtered tokens
filtered_tokens = []

# Iterate over each sentence
for sentence in mis_postag_and_ner_labels_list:
    # Filtered tokens based on POS tags for the current sentence
    tokens = [token[0] for token in sentence if token[1] in filtered_pos_tags]
    filtered_tokens.append(' '.join(tokens))

# Calculate TF-IDF scores
vectorizer = TfidfVectorizer()
tfidf_scores = vectorizer.fit_transform(filtered_tokens)

# Get the feature names (tokens)
feature_names = vectorizer.get_feature_names()

# List to store the top tokens and their TF-IDF scores
top_tokens = []
top_scores = []

# Iterate over each sentence's TF-IDF scores
for i, sentence_scores in enumerate(tfidf_scores):
    # Get the TF-IDF scores for the current sentence
    scores = sentence_scores.toarray()[0]
    
    # Combine the tokens with their TF-IDF scores
    token_scores = zip(feature_names, scores)
    
    # Sort the tokens based on their TF-IDF scores in descending order
    sorted_tokens = sorted(token_scores, key=lambda x: x[1], reverse=True)
    
    # Select the top tokens and their TF-IDF scores
    top_tokens.extend([token for token, _ in sorted_tokens[:200]])
    top_scores.extend([score for _, score in sorted_tokens[:200]])

# Generate word cloud
wordcloud_data = {token: score for token, score in zip(top_tokens, top_scores)}
wordcloud = WordCloud(width=1200, height=600, background_color='white').generate_from_frequencies(wordcloud_data)

# Plot the word cloud
plt.figure(figsize=(15, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
