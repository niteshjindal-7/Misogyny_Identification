```python
import pandas as pd
import os
import numpy as np
import scipy.stats
```

```python
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h, h
```

```python
# Directory path containing the Excel files
directory = '/home/nitesh/Documents/MY_THESIS/MAMI'

# List to store individual DataFrames
dfs = []

# Iterate over each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.xlsx'):
        # Construct the file path
        file_path = os.path.join(directory, filename)
        
        # Read the Excel file into a DataFrame
        df = pd.read_excel(file_path)
        
        columns = ['Model', 'Accuracy', 'M-Precision', 'M-Recall', 'M-F1-Score',
           'W-Precision', 'W-Recall', 'W-F1-Score']

        df1 = pd.DataFrame(columns=columns)

        row_data = {

        'Model': df.Model[0],

        'Accuracy': mean_confidence_interval(df['Accuracy'], confidence=0.95)[0],
        'Accuracy_CI': mean_confidence_interval(df['Accuracy'], confidence=0.95)[3],

        'M-Precision': mean_confidence_interval(df['M-Precision'], confidence=0.95)[0],
        'M-Precision_CI': mean_confidence_interval(df['M-Precision'], confidence=0.95)[3],

        'M-Recall': mean_confidence_interval(df['M-Recall'], confidence=0.95)[0],
        'M-Recall_CI': mean_confidence_interval(df['M-Recall'], confidence=0.95)[3],

        'M-F1-Score': mean_confidence_interval(df['M-F1-Score'], confidence=0.95)[0],
        'M-F1-Score_CI': mean_confidence_interval(df['M-F1-Score'], confidence=0.95)[3],

        'W-Precision': mean_confidence_interval(df['W-Precision'], confidence=0.95)[0],
        'W-Precision_CI': mean_confidence_interval(df['W-Precision'], confidence=0.95)[3],

        'W-Recall': mean_confidence_interval(df['W-Recall'], confidence=0.95)[0],
        'W-Recall_CI': mean_confidence_interval(df['W-Recall'], confidence=0.95)[3],

        'W-F1-Score': mean_confidence_interval(df['W-F1-Score'], confidence=0.95)[0],
        'W-F1-Score_CI': mean_confidence_interval(df['W-F1-Score'], confidence=0.95)[3],

#         'Runtime': mean_confidence_interval(df['Runtime'], confidence=0.95)[0],
#         'Runtime_CI': mean_confidence_interval(df['Runtime'], confidence=0.95)[3]
        
        }

        df1 = pd.concat([df1, pd.DataFrame([row_data])], ignore_index=True)
 
        # Append the DataFrame to the list
        dfs.append(df1)

# # Concatenate the DataFrames row-wise
concatenated_df = pd.concat(dfs, axis=0)
# pd.DataFrame(dfs)
# # # Reset the index of the concatenated DataFrame
concatenated_df = concatenated_df.reset_index(drop=True)

# # # Display the concatenated DataFrame
display(concatenated_df)
```
